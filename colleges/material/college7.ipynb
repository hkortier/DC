Voor de parallel schakeling (links) geldt:
$$L_{eq} = \dfrac{1}{\dfrac{1}{L_1}+\dfrac{1}{L_2}} = \dfrac{L_1 L_2}{L_1 + L_2}$$

Voor de serie schakeling (rechts) geldt:
$$L_{eq} = L_3 + L_4$$

### Energie in een spoel

De spoel slaat energie op in het Magnetische veld. 

De arbeid die men moet verrichten om een lading te verplaatsen over een spannings (potentiaal) V:

$ U = \displaystyle\int^I_0 LI dI $


Dit komt aan bod in **trillingen en golven** in het tweede jaar. De uitwerking van de integraal is:

\begin{equation}
E_{mag,pot} = \frac{1}{2} \cdot L \cdot I^2
\end{equation}

Herken hier de overeenkomst met de energie opslag in een condensator:

\begin{equation}
E_{el,pot} = \frac{1}{2} \cdot C \cdot V^2
\end{equation}

### Oplaad en ontlaad functie

De spanning over een spoel wordt bepaald door de verandering van stroom door de spoel vermedigvuldig met de zelfinductie:

$u_l = L \dfrac{d i_l}{dt}$

Het opladen en ontladen kost een zekere tijd.

De tijdconstante is een vast tijdsinterval dat gelijk is aan het quotient van zelfinductie en weerstand van een RL circuit in serie:

$\tau = \dfrac{L}{R}~[s]$

Op $t = 5\tau$ is de spoel 99% geladen. 

In het geval van een RL netwerk levert dit de volgende vergelijkingen voor het op en ontladen:


### Algemene vergelijking:

\begin{equation}
i_l (t) = I_f + \left( I_i - I_f \right) \cdot e^{\left( \dfrac{-t}{\tau} \right)}
\end{equation}

met $I_i$ en $I_f$ de initiele (initial) en uiteindelijke (final) stromen zijn.

#### Opladen vanaf 0 Ampere ($I_i=0$):

$i_l (t) = I_{f} \cdot \left( 1 - e^{\dfrac{-t}{\tau}} \right)$

waarbij $I_f$ de eindstroom is.

#### Ontladen NAAR 0 Ampere ($I_f=0$)

$i_l (t) = I_{i} \cdot e^{\dfrac{-t}{\tau}}$

waarbij $I_i$ de initiele stroom is.

------

```python
t = np.linspace(0,5,100)
R = 220
L = 220
I0 = 10

tau = L/R

il_op = I0*(1-np.exp(-t/tau)) #opladen
il_ont = I0*np.exp(-t/tau) #ontladen
plt.plot(t,il_op,t,il_ont)
plt.xlabel('Tijd [s]')
plt.ylabel('Stroom [A]')
```

Stel het volgende schakelpatroon voor:

```python
fs = 100
tEnd = 6
N = fs*tEnd

time = np.linspace(0,tEnd,N)
t = np.tile(np.linspace(0,1,fs), tEnd) #switching every second
squareInputSignal = np.tile(np.r_[np.ones(100), np.zeros(100)], tEnd/2)

R = 1
L = 2
tau = L/R

IIstart = 0 #initial voltage
IFstart = 2 #final voltage

II = IIstart
IF = IFstart
prevState = 1

iOut = np.zeros(N)

for i in range(N):
    if squareInputSignal[i] == 0:
        if prevState == 1:
            IF = IIstart
            II = iOut[i-1]
        prevState = 0
    else:
        if prevState == 0:
            IF = IFstart
            II = iOut[i-1]
        prevState = 1    
    
    iOut[i] = IF + (II-IF)*np.exp(-t[i]/tau)
        
plt.figure(num=None, figsize=(12, 6))
plt.plot(time,squareInputSignal)        
plt.plot(time,iOut)
plt.xlabel('Tijd [s]')
plt.ylabel('Stroom i_l~[A]')
```
